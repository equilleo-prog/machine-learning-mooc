Imagine we are interested in predicting penguins species based on two of their
body measurements: culmen length and culmen depth. First we want to do some
data exploration to get a feel for the data.

```python
import pandas as pd
import matplotlib
import seaborn as sns

penguins = pd.read_csv("../datasets/penguins_classification.csv")

penguins.head() # show first rows of the data
target_column = "Species"
penguins[target_column].value_counts()
```

What are the features?
<span style="color:green">= culmen length and depth</span>  
What is the target?
<span style="color:green">= penguin specy, classification problem</span>  
How many features are numerical? How many features are categorical?
<span style="color:green">= all (2) features are numerical and the target is categorical</span>  
What are the different penguins species available in the dataset and how many
samples of each species are there? <span style="color:green">= 3 different species, Adelie : 151, Gentoo : 123, Chinstrap : 68</span> 

```python
_ = penguins.hist(figsize=(8, 4)) # plot histograms
```

```python
columns = penguins.columns[0:-1]
target_column = penguins.columns[-1] # show features distribution for each class
_ = sns.pairplot(
    data=penguins, # DataFrame containing the data
    vars=columns, # List of columns to include in the pairplot
    hue=target_column, # Column name for color coding the target
    plot_kws={"alpha": 0.2}, # Setting transparency to 0.2
    height=3, # Height of each facet in the pairplot
    diag_kind="hist", # Type of plot for the diagonal subplots, set to histogram
    diag_kws={"bins": 30}, # Setting the number of bins in the histogram to 30
)
```

Looking at these distributions, how hard do you think it would be to classify
